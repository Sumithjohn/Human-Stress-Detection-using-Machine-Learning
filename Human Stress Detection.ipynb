```python
x=df.drop(['Stress Levels'],axis=1)
x
```

```python
y=df['Stress Levels']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
l_model=LogisticRegression()
```

```python
l_model.fit(X_train,y_train)
```

```python
y_pred=l_model.predict(X_test)
y_pred
```

```python
accuracy=l_model.score(X_test,y_test)
print('L_accuracy :',accuracy)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.title('Confusion Matrix for LogisticRegression ')
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=100)
```

```python
rf_model.fit(X_train,y_train)
```

```python
accuracy1=rf_model.score(X_test,y_test)
print('Rf_accuracy :',accuracy1)
```

```python
y_pred_rf=rf_model.predict(X_test)
y_pred_rf
```

```python
rf_cm=confusion_matrix(y_test,y_pred_rf)
rf_cm
```

```python
sns.heatmap(rf_cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.title('Confusion Matrix for Random Forest ')
plt.show()
```

```python
print(classification_report(y_test, y_pred_rf))
```

# KNeighbors

```python
knn_model.score(X_test,y_test)
```

```python
y_pred_k=knn_model.predict(X_test)
y_pred_k
```

```python
knn_cm=confusion_matrix(y_test,y_pred_k)
knn_cm
```

```python
sns.heatmap(knn_cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('Truth')
plt.title('Confusion Matrix for knn')
plt.show()
```

```python
print(classification_report(y_test,y_pred_k))
```
